```python
#We are going to import the libraries that we need and our file, with one sinlge column for the moment.

import re
import pandas as pd
pd.set_option('display.max_colwidth', None)
df = pd.read_csv('Web_access_log-akumenius.com.txt',names=["unica_columna"])
df.sample(5)
```

```python
#Lets split our column by the following patron " ", and create a dataframe.
df = df['unica_columna'].str.split(' ', expand=True)
df.sample(5)
```

```python
#By indexing our columns now we can create and define each one, to start cleaning the data.
df['web']= df[0]
df['ip']= df[1]
df['date']= df[4]
df['get']= df[6]+df[7]+df[8]
df['get'] = df['get'].str[1:-1]

#Lets clean our date column by using split and replace mode.
df[["day", "month", "year"]] = df["date"].str.split("/", expand = True)
df['day'] = df['day'].str.replace('[', " ", regex=True)
df['year'] = df['year'].replace('', "", regex=True)
df['year'] = df['year'].str[:-9]
df['hour'] = df['date'].str[13:]

df['numbers'] = df[9]
df.head()

#Notice that we use some parameters to clean our data such a regex epressions or split parameters that we can use when manipulate strings.
```

```python
#Lets create a new dataframe with the new columns.

df2 = df[['web', 'ip', 'get', 'numbers', 'day', 'month', 'year', 'hour']]
df2.head()
```

```python
#Creating a DataFrame with the numbers of visits depending on the ip.
visitas= df2["ip"].value_counts().rename_axis('ip').reset_index(name="visitas")
visitas.head(9)
```
